## **iv. Data Cleaning**

Convert Launch Date to Datetime format to ensure concistency and facilitate time-based analysis for easier filtering, sorting and time-series analysis of the token launches.

```python
df['Launch_Date'] = pd.to_datetime(df['Launch Date'])
```

Filtering data only for token launched in 2023 to ensure that the analysis focuses on recent market trends, the dataset is filtered to include only meme coin launched in 2023. This help the relevance evaluation of trading activity and liquidity trends withtin the most recent time frame.

```python
df_2023 = df[df['Launch_Date'].dt.year == 2023]
print(f"Total rows for 2023: {df_2023.shape[0]}")
```

Random sampling up to 200 Meme Coins to ensure a manageable and representative subset for analysis, 200 meme coins from the 2023 dataset are randomly selected. This approach helps in maintain unbiased selection while still preserve the overall market trends within the data.

```python
sampled_df = df_2023.sample(n=200, random_state=42)
print(sampled_df.shape)
```

Replace spaces on column names to "_" for simplicity and consistency.

```python
sampled_df.rename(columns=lambda x: x.strip().replace(" ", "_"), inplace=True)
sampled_df
```

Drop the symbol column for cleaner data.

```python
sampled_df = sampled_df.drop(columns=['Symbol'])
sampled_df
```

We decided to drop the symbol name because the value data is mismatch with the actual ticker or name of the token. So for a cleaner analysis we decided to drop it since the value could already be represented by column's Coin_Name.

Next, is to analyze the correlation between Market Cap and Trading and Volume using a scatter plot to help visualize wether higher Market Cap is associated with higher Trading Volume, which is crucial in understanding the liquidity trends among meme coins.

```python
# Scatter Plot
sns.scatterplot(x='Market_Cap_(USD)', y='Trading_Volume_(24h)', data=sampled_df)
plt.title("Market Cap vs Trading Volume")
plt.xlabel("Market Cap")
plt.ylabel("Trading Volume")
plt.show()
```

Insights:

- The points appear widely dispersed which indicates that a higher Market Cap does not necessarily result in higher Trading Volume.

- We foudn that Some meme coins with lower Market Cap still exhibit high trading volume, suggesting that liquidity is not influenced solely by Market Cap which indicates other variables like hype, community engagement or Centralized Exchange (CEX) listings could influence the liquidity.

Next is to conduct a Pearson Correlation test to determine the significane of correlation among Market Cap and Trading Volume.

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(sampled_df['Market_Cap_(USD)'], sampled_df['Trading_Volume_(24h)'])
print(f"Pearson Correlation: {corr}, P-value: {p_value}")
```

Insight :

- The Pearson Correlation Coefficient (0.0397) indicates a very weak relationship between Market Cap and Trading Volume. This suggests that higher Market Cap does not necessarily correspond to higher trading activity.

- The P-value (0.5771) is much higher than the standard significance level (0.05), meaning the observed correlation is not statistically significant.

These results confirm that Market Cap alone is not a strong predictor of Trading Volume, and other factors like hype, community interest, or external influences may play a larger impact in meme coin liquidity.

Since the statistical correlation calculation showing weak relationship among variables we decide to add Spearman Correlation test the monotonic relationship and add derived metrics called turnover ratio(Trading Volume/Market Cap) to explore more about the liqudity trends.

```python
# Add Turnover Ratio

sampled_df['Turnover_Ratio'] = sampled_df['Trading_Volume_(24h)'] / sampled_df['Market_Cap_(USD)']

#Visual
sampled_df['Turnover_Ratio'].hist(bins=20, color='pink')
plt.title("Distribution of Turnover Ratio")
plt.xlabel("Turnover Ratio")
plt.ylabel("Frequency")
plt.show()
```

Insight:

- The Turnover Ratio (Trading Volume / Market Cap) is heavily right-skewed, meaning that most meme coins have a low turnover ratio, while only a few exhibit high liquidity relative to their market cap.

- A few outliers exceed 0.5, meaning certain meme coins are being traded at half or more of their market cap within 24 hours, which may indicate speculative trading or short-term hype-driven activity.

- The sharp drop in frequency as Turnover Ratio increases suggests that highly liquid meme coins are rare, reinforcing the idea that Market Cap alone is not a strong predictor of liquidity.

This analysis revealed that most meme coins on solana have low liquidity with only a few exceptions exhibiting high trading activity relative to their market cap. The findings support that the Market Cap alone can't be a predictor of liquidity on this market as highly liquid coins are often associated with short-term speculation and hype-driven trading.

This findings emphasize the need for investors and traders to consider additional liquidity indicators like trends or other external sentiment factors.

Next, we are going to highlight the top 10 of meme coins with the consistent high Turnover Ratios that can be an indicator of a much safer assets to trade so the user will have less risk on getting rug or pump and dump schemes.

```python
top_turnover = sampled_df.sort_values(by='Turnover_Ratio', ascending=False).head(10)
print("Top 10 Coins by Turnover Ratio:")
print(top_turnover[['Coin_Name', 'Turnover_Ratio']])
```

```python
# Visuals
sns.barplot(x='Turnover_Ratio', y='Coin_Name', data=top_turnover, palette='viridis')
plt.title("Top 10 Coins by Turnover Ratio")
plt.xlabel("Turnover Ratio")
plt.ylabel("Coin Name")
plt.show()
```

Spearman Correlation Test:   
Since the Pearson Correlation Test showed a weak relationship between Market Cap and Trading Volume, we add a Spearman Correlation to determine whether an increase in Market Cap consistently corresponds to an increase in Trading Volume, even if the relationship is not strictly linear.  

If the Spearman Correlation is strong it's indicate that Market Cap and Trading Volume move together in a ranked order, even if the relationship is not strictly linear. However, if the correlation remains weak then it further confirms that Market Cap alone cannot determine trading activity.


```python
from scipy.stats import spearmanr

# Calculate Spearman Correlation
spearman_corr, spearman_p = spearmanr(sampled_df['Market_Cap_(USD)'], sampled_df['Trading_Volume_(24h)'])
print(f"Spearman Correlation: {spearman_corr:.4f}, P-value: {spearman_p:.4f}")

# Interpretation
if spearman_p < 0.05:
    print("The Spearman correlation is statistically significant.")
else:
    print("The Spearman correlation is not statistically significant.")
```

Insights:  
- The **Spearman Correlation Coefficient (0.0318)** indicates a **very weak monotonic relationship** between **Market Cap and Trading Volume**.  
- The **P-value (0.6548)** is **greater than 0.05**, meaning the correlation **is not statistically significant** and there is **no meaningful pattern** in how Market Cap and Trading Volume are ranked.  
- These findings align with the **Pearson Correlation results**, reinforcing the idea that Market Cap does not strongly influence Trading Volume in Solana meme coins.
